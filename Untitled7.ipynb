```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df_king = pd.read_csv("kca_house_data.csv")

# Drop duplicate rows
df_king.drop_duplicates(inplace=True)

# Remove unwanted columns
columns_to_drop = ['id', 'date', 'sqft_living15', 'sqft_lot15', 'waterfront', 
                   'view', 'sqft_above', 'sqft_basement', 'yr_renovated', 
                   'lat', 'long', 'zipcode']
df_king.drop(columns_to_drop, axis=1, inplace=True)

# Handling missing values
# For numerical columns, fill missing values with mean
numerical_columns = df_king.select_dtypes(include=['float64', 'int64']).columns
df_king[numerical_columns] = df_king[numerical_columns].fillna(df_king[numerical_columns].mean())
# Define a function to remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Remove outliers for each numerical column
for column in numerical_columns:
    df_king = remove_outliers_iqr(df_king, column)

# Display the updated DataFrame
print(df_king.head())


# Plot scatter plots to check linearity
target_variable = 'price'  # Change this to your target variable
for column in df_king.columns:
    if column != target_variable:
        plt.figure(figsize=(8, 6))
        plt.scatter(df_king[column], df_king[target_variable], alpha=0.5)
        plt.title(f'Scatter plot of {column} vs {target_variable}')
        plt.xlabel(column)
        plt.ylabel(target_variable)
        plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the data
df_king = pd.read_csv("kca_house_data.csv")

# Drop duplicate rows
df_king.drop_duplicates(inplace=True)

# Remove unwanted columns
columns_to_drop = ['id', 'date', 'sqft_living15', 'sqft_lot15', 'waterfront', 
                   'view', 'sqft_above', 'sqft_basement', 'yr_renovated', 
                   'lat', 'long', 'zipcode']
df_king.drop(columns_to_drop, axis=1, inplace=True)

# Handling missing values
# For numerical columns, fill missing values with mean
numerical_columns = df_king.select_dtypes(include=['float64', 'int64']).columns
df_king[numerical_columns] = df_king[numerical_columns].fillna(df_king[numerical_columns].mean())

# Define target variable and features
target_variable = 'price'
features = [col for col in df_king.columns if col != target_variable]

# Remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# Remove outliers for each numerical column
for column in numerical_columns:
    df_king = remove_outliers_iqr(df_king, column)

# Add intercept term
X = sm.add_constant(df_king[features])
y = df_king[target_variable]

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
```

The R-squared value of 0.517 indicates that approximately 51.7% of the variability in house prices can be explained by the independent variables included in the model. Among the coefficients, 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'condition', 'grade', and 'yr_built' are statistically significant predictors of house prices, as their p-values are less than 0.05. However, the coefficient for 'bedrooms' is not statistically significant. Violations of model assumptions include non-normality of residuals (as indicated by the significant p-values in the Omnibus and Jarque-Bera tests) and potential multicollinearity issues (as suggested by the large condition number). These violations indicate that the model may not fully meet the assumptions of linear regression, and further diagnostics and improvements may be necessary to enhance its reliability and predictive performance.

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Assuming you already have X, y, and model defined as in the previous code

# Calculate residuals
residuals = model.resid

# Check for Normality of Residuals
# Q-Q plot
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals")
plt.show()

# Histogram of residuals
plt.hist(residuals, bins=20)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()

# Check for Homoscedasticity
# Plot residuals against fitted values
plt.scatter(model.fittedvalues, residuals, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.title("Residuals vs Fitted Values")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

# Alternatively, you can plot residuals against each independent variable
for column in X.columns:
    if column != 'const':
        plt.scatter(X[column], residuals, alpha=0.5)
        plt.axhline(y=0, color='r', linestyle='--')
        plt.title(f"Residuals vs {column}")
        plt.xlabel(column)
        plt.ylabel("Residuals")
        plt.show()
```

The R-squared value of 0.329 indicates that approximately 32.9% of the variance in the house prices can be explained by the independent variables included in the model. Among the coefficients, 'sqft_living' and 'sqft_lot' are statistically significant predictors of price, with coefficients of 189.7418 and -10.5594 respectively. However, some coefficients, such as those for certain bathroom features like 'bath_0.75' and 'bath_1.25', have high p-values, suggesting they may not significantly contribute to explaining the variation in house prices. Additionally, the model's assumptions may be violated, as indicated by the statistically significant Omnibus test, which suggests that the residuals are not normally distributed. This violation could affect the reliability of the coefficient estimates and hypothesis tests. Further diagnostics and potentially refining the model by addressing multicollinearity and non-normality of residuals may improve its performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Instantiate the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on training and testing sets
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluate the model
train_rmse = mean_squared_error(y_train, train_preds, squared=False)
test_rmse = mean_squared_error(y_test, test_preds, squared=False)
train_r2 = r2_score(y_train, train_preds)
test_r2 = r2_score(y_test, test_preds)

print("Training set RMSE:", train_rmse)
print("Testing set RMSE:", test_rmse)
print("Training set R-squared:", train_r2)
print("Testing set R-squared:", test_r2)

# Check if it's underfitting or overfitting
if train_r2 < test_r2:
    print("The model is potentially overfitting.")
elif train_r2 > test_r2:
    print("The model is potentially underfitting.")
else:
    print("The model is performing similarly on both training and testing sets.")
```
